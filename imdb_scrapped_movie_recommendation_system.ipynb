```python
movies_df = imdb_df[['MovieName', 'Storyline', 'Genre', 'MPAA', 'TopCast', 'Plot_Keywords', 'Director']]
```

```python
movies_df['TopCast'][0]
```

```python
import ast
def getcast(obj):
    cast = []
    x = 0
    for i in ast.literal_eval(obj):
        if(x < 6):
            cast.append(i['TopCast'])
            x = x+1
        else:
            break
    return cast
```

```python
movies_df['TopCast'].apply(getcast)
```

```python
movies_df['TopCast'] = movies_df['TopCast'].apply(getcast)
```

```python
movies_df.head()
```

```python
movies_df['TopCast'] = movies_df['TopCast'].apply(lambda x:" ".join(x))
```

```python
movies_df.head()
```

```python
def getkeywords(obj):
    keywords = []
    x = 0
    try:
        for i in ast.literal_eval(obj):
            if(x < 6):
                keywords.append(i['Plot_Keywords'])
                x = x+1
            else:
                break
    except:
        keywords.append("movie")
    return keywords
```

```python
movies_df['Plot_Keywords'] = movies_df['Plot_Keywords'].apply(getkeywords)
```

```python
movies_df['Plot_Keywords'] = movies_df['Plot_Keywords'].apply(lambda x:" ".join(x))
```

```python
movies_df.head()
```

```python
movies_df['features'] = movies_df['Storyline'] + movies_df['TopCast'] + movies_df["Plot_Keywords"] + movies_df["Director"]
```

```python
final_movies_df = movies_df[['MovieName', 'features']]
```

```python
final_movies_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 4000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(final_movies_df['features']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
final_movies_df['features'] = final_movies_df['features'].apply(stem)
```

```python
vectors = cv.fit_transform(final_movies_df['features']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = final_movies_df[final_movies_df['MovieName'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(final_movies_df.iloc[i[0]].MovieName)
```

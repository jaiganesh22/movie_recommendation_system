```python
import numpy as np
import pandas as pd
from scrapy.selector import Selector
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
movies_list = []
movies_tomatometer = []
movies_info = []
movies_genre = []
movies_critics = []
movies_critics_ratings = []

```

```python
for i in range(1,4):
    driver = webdriver.Chrome('chromedriver.exe')
    url = f'https://www.rottentomatoes.com/browse/movies_in_theaters/sort:newest?page={i}'
    time.sleep(1)
    driver.get(url)
    time.sleep(1)
    print(driver.title)
    time.sleep(1)
    body = driver.find_element(By.CSS_SELECTOR, 'body')
    
    sel = Selector(text = driver.page_source)
    movies_links = sel.css('.discovery-tiles .discovery-tiles__wrap a::attr(href)').extract()
    
    for i in range(len(movies_links)):
        movie = movies_links[i]
        movie = movie.replace('/m/','')
        movie = movie.replace('_', ' ')
        movies_list.append(movie)
        
    
    for movie in movies_links:
        req_link = f"//a[@href='{movie}']"
        try:
            time.sleep(1)
            movie_link = driver.find_element(By.XPATH, req_link)
            movie_link.click()
            time.sleep(1)
            try:
                time.sleep(1)
                movie_score_element = driver.find_elements(By.CSS_SELECTOR, 'div#topSection div.thumbnail-scoreboard-wrap')
                time.sleep(1)
                print(movie_score_element)
                movie_score_element = movie_score_element[0]
                score_sel = Selector(text = movie_score_element.get_attribute('innerHTML'))
                movie_score = score_sel.css('score-board::attr(tomatometerscore)').extract()[0]
                movies_tomatometer.append(movie_score)
            except:
                movies_tomatometer.append('')
    
            print(driver.current_url)
            try:
                time.sleep(1)
                movie_info_element = driver.find_elements(By.CSS_SELECTOR, "div.media-body div.panel-body div#movieSynopsis")[0]
                time.sleep(1)
                info_sel = Selector(text = movie_info_element.get_attribute('innerHTML'))
                movie_info = info_sel.extract()
                movies_info.append(movie_info)
        
            except:
                movies_info.append('')
    
            try:
                time.sleep(1)
                content_element = driver.find_elements(By.CSS_SELECTOR, "ul.content-meta.info ")[0]
                time.sleep(1)
                content_selector = Selector(text = content_element.get_attribute('innerHTML'))
                content_list = content_selector.css('.meta-value::text').extract()
                genre = content_list[0]
                movies_genre.append(genre)
        
            except:
                movies_ratings.append('')
                movies_genre.append('')
    
            try:
                time.sleep(1)
                reviews_element = driver.find_elements(By.CSS_SELECTOR, "div.reviews-wrap")[0]
                time.sleep(1)
                reviews_sel = Selector(text = reviews_element.get_attribute('innerHTML'))
                cur_reviews = reviews_sel.css('review-speech-balloon::attr(reviewquote)').extract()
                movies_critics.append(cur_reviews)
                cur_ratings = reviews_sel.css('review-speech-balloon::attr(originalscore)').extract()
                movies_critics_ratings.append(cur_ratings)
        
            except:
                movies_critics.append('')
                movies_critics_ratings.append('')
    
            driver.back()
        except:
            pass
```

```python
movies_df = pd.DataFrame(list(zip(movies_list, movies_tomatometer, movies_info, movies_genre, 
                        movies_critics, movies_critics_ratings)),
                        columns=['Movie_name','Tomatometer','MovieInfo','Genre','Critics','CriticsRatings'])
```
